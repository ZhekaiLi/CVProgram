```python
import numpy as np
import open3d as o3d
```

```python
pcd=o3d.io.read_triangle_mesh("cantilever.obj")
print(pcd.vertices)
point = np.asarray(pcd.vertices)  # 将点转换为numpy数组
# path为文件路径
```

```python
hull, _ = pcd.compute_convex_hull()
hull_ls = o3d.geometry.LineSet.create_from_triangle_mesh(hull)
hull_ls.paint_uniform_color((1, 0, 0))
o3d.visualization.draw_geometries([pcd, hull_ls])
```

```python
print(hull,_)
point2=np.asarray(hull.vertices)
print(point2)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(x=point2[:,0], y=point2[:,1], z=point2[:,2],
                                   mode='markers',
                                   marker=dict(
                                    size=5,
                                                    # set color to an array/list of desired values
                                    colorscale='Viridis',   # choose a colorscale
                                    opacity=0.8
                                     ))])

fig.show()
```

```python
import matplotlib.pyplot as mp
x=point2[:,0]
y=point2[:,1]
z=point2[:,2]

# 2.绘制图片
mp.figure("3D Scatter", facecolor="lightgray")
ax3d = mp.gca(projection="3d")  # 创建三维坐标

mp.title('possible points', fontsize=20)
ax3d.set_xlabel('x', fontsize=14)
ax3d.set_ylabel('y', fontsize=14)
ax3d.set_zlabel('z', fontsize=14)
mp.tick_params(labelsize=10)



ax3d.scatter(x,y,z, s=20,  cmap="jet", marker="o")

mp.show()

```

```python
n=x.size
x_left=[]
y_left=[]
z_left=[]
x_right=[]
y_right=[]
z_right=[]
for i in range(n):
    if x[i]<-10:
        x_left.append(x[i])
        y_left.append(y[i])
        z_left.append(z[i])
    else:
        x_right.append(x[i])
        y_right.append(y[i])
        z_right.append(z[i])
    
```

```python
mp.figure("3D Scatter", facecolor="lightgray")
ax3d = mp.gca(projection="3d")  # 创建三维坐标

mp.title('left points', fontsize=20)
ax3d.set_xlabel('x', fontsize=14)
ax3d.set_ylabel('y', fontsize=14)
ax3d.set_zlabel('z', fontsize=14)
mp.tick_params(labelsize=10)



ax3d.view_init(elev=0,    # 仰角
             azim=0       # 方位角
              )


ax3d.scatter(x_left,y_left,z_left, s=20,  cmap="jet", marker="o")


mp.show()

```

```python
mp.subplot(221)
mp.scatter(x_left,y_left, s=20,  cmap="jet", marker="o")

mp.title("x-y") #图名
mp.xlabel("x")#x轴标签
mp.ylabel("y")#y轴标签
mp.tick_params(axis='both')#x,y轴都有刻度

mp.subplot(222)
mp.scatter(x_left,z_left, s=20,  cmap="jet", marker="o")

mp.title("x-z") #图名
mp.xlabel("x")#x轴标签
mp.ylabel("z")#y轴标签
mp.tick_params(axis='both')#x,y轴都有刻度

mp.subplot(223)
mp.scatter(y_left,z_left, s=20,  cmap="jet", marker="o")

mp.title("y-z") #图名
mp.xlabel("y")#x轴标签
mp.ylabel("z")#y轴标签
mp.tick_params(axis='both')#x,y轴都有刻度

mp.show
```

```python
mp.figure("3D Scatter", facecolor="lightgray")
ax3d = mp.gca(projection="3d")  # 创建三维坐标

mp.title('left points', fontsize=20)
ax3d.set_xlabel('x', fontsize=14)
ax3d.set_ylabel('y', fontsize=14)
ax3d.set_zlabel('z', fontsize=14)
mp.tick_params(labelsize=10)



ax3d.view_init(elev=0,    # 仰角
             azim=30       # 方位角
              )


ax3d.scatter(x_left,y_left,z_left, s=20,  cmap="jet", marker="o")

xl=[-21.2,-21.2,-21.2,-21.2]
yl=[-0.6,-0.8,-1.37,-1.1]
zl=[-5.75,-5.9,-5.25,-5.14]
ax3d.scatter(xl,yl,zl, s=20,  cmap='red', marker="o")

mp.show()
```

```python
mp.figure("3D Scatter", facecolor="lightgray")
ax3d = mp.gca(projection="3d")  # 创建三维坐标

mp.title('right points', fontsize=20)
ax3d.set_xlabel('x', fontsize=14)
ax3d.set_ylabel('y', fontsize=14)
ax3d.set_zlabel('z', fontsize=14)
mp.tick_params(labelsize=10)

ax3d.view_init(elev=10,    # 仰角
             azim=-10       # 方位角
              )


ax3d.scatter(x_right,y_right,z_right, s=20,  cmap="jet", marker="o")


mp.show()
```
