```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r1 = np.linspace(0, 30, 100)
r2 = np.linspace(30.01, 40, 100)
r3 = np.linspace(40.01, 100, 100)
r4 = np.linspace(100.01, 30*np.sqrt(17), 100)
r5 = np.linspace(30*np.sqrt(17)+0.01, 40*np.sqrt(10), 100)
```

```python
y1 = 5/8 * r1/5000
y2 = 5/8 * r2/5000 + 1/8 * r2/(120*np.sqrt(r2**2-900))
y3 = 5/8 * r3/5000 + 1/8 * r3/(120*np.sqrt(r3**2-900)) + 3/8 * r3/(120*np.sqrt(r3**2-1600))
y4 = 1/8 * r4/(120*np.sqrt(r4**2-900)) + 3/8 * r4/(120*np.sqrt(r4**2-1600))
y5 = 3/8 * r5/(120*np.sqrt(r5**2-1600))
```

```python
rall = np.append(r1, [r2, r3, r4, r5])
yall = np.append(y1, [y2, y3, y4, y5])
```

```python
plt.plot(rall, yall)
plt.savefig('3.jpg')
```

```python
a1 = np.linspace(0, 0.2, 100)
a2 = np.linspace(0.2001, 1.2, 100)
```

```python
y11 = 10*np.exp(-10*a1)
y12 = 10*np.exp(-10*a2)

y21 = 1838.48*a1**2*np.exp(-10*a1)
y22 = 73.54*np.exp(-10*a2)

y31 = 11.574*(1-25*a1**2)*np.exp(-10*a1)
```

```python
a = np.append(a1, a2)
y1 = np.append(y11, y12)
y2 = np.append(y21, y22)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(a, y1, label='f(a)')
plt.plot(a, y2, label='f(a|D)')
plt.plot(a1, y31, label='f(a|not D)')
plt.legend()
plt.show()
plt.savefig('2c.jpg')
```
