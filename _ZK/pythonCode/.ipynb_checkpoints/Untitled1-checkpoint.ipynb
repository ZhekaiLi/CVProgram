```python
import trimesh
import numpy as np
import sympy as sy
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from SkeletonOfBeam import SkeletonOfBeam
from SkeletonOfBeam import GeometryToolBox
from SkeletonOfBeam import SegsOfSOB
```

```python
mesh = trimesh.load_mesh("cantilever.stl")
```

```python
sob = SkeletonOfBeam(mesh, [1, 0, 0])
sob.getScaleAlongSkeletonVec()
sob.getIntersectionsFromStep(step=1)
sob.getSkeletonPoints()
sob.getNewCoordinate()
sob.getProjections()
sob.getSkeletonEqs()
sob.getDerivativeSkeletonEqs()
sob.getNewSkeletonPoints()
sob.getNewIntersections()
```

```python
sos = SegsOfSOB(sob, 10)
sos.getNewA()
sos.getNewEqs()
sos.getNewDerivativeEqs()
sos.getSobs()
```

```python
sos.nSobs[-1].showIntersections()
```

#### Avg Cross-section Area

```python
sum([sec.area for sec in sos.nSobs[-1].Intersections2D]) / len(sos.nSobs[-1].Intersections2D)
```

```python
np.std([sec.area for sec in sos.nSobs[-1].Intersections2D])
```
