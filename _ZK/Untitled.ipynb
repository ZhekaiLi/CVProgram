```python
fig = plt.figure()  #
ax3 = plt.axes(projection='3d')

#
xx = np.arange(50.5,np.sqrt(50**2+190**2),1)
yy = np.arange(10,100,1)
X, Y = np.meshgrid(xx, yy)

for i in range(len(Y)):
    valY = Y[i][0]
    maxX = np.sqrt(50**2 + (200-valY)**2)
    for j in range(len(Y[0])):
        if X[i][j] >= maxX:
            X[i][j] = 0   
            
        
Z = 100/9*X / (Y**2*(200-Y)*np.sqrt(X**2-50**2))

for i in range(len(Y)):
    for j in range(len(Y[0])):
        if np.isnan(Z[i][j]):
            Z[i][j] = 0   
            
#
ax3.plot_surface(X,Y,Z, rstride = 1, cstride = 1,cmap='rainbow')
#ax3.contour(X,Y,Z, zdim='z',offset=-2cmap='rainbow)   #offsetZ
plt.show()
```
