```python
import open3d as o3d
from open3d.web_visualizer import draw
```

```python
cube_red = o3d.geometry.TriangleMesh.create_box(1, 2, 4)
cube_red.compute_vertex_normals()
cube_red.paint_uniform_color((1.0, 0.0, 0.0))
draw(cube_red)
```

```python
# obj面片显示
textured_mesh= o3d.io.read_triangle_mesh('cantilever.obj')
print(textured_mesh)
textured_mesh.compute_vertex_normals()
```

```python
pcobj = o3d.geometry.PointCloud()
pcobj.points = o3d.utility.Vector3dVector(textured_mesh.vertices)
```

```python
draw(pcobj.points)
```

```python
canti = o3d.geometry.TriangleMesh.create_from_point_cloud_alpha_shape(pcobj, 0.5)
canti.compute_vertex_normals()
canti.paint_uniform_color((1.0, 0.0, 0.0))
draw(canti)
```

```python
mesh_simp = canti.simplify_quadric_decimation(target_number_of_triangles=5000)
print(mesh_simp)
```

```python
mesh_simp.compute_vertex_normals()
mesh_simp.paint_uniform_color((1.0, 0.0, 0.0))
draw(mesh_simp)
```
